```python
import google.generativeai as genai
```

```python
f = open("keys/gemini.txt")
key = f.read()

genai.configure(api_key=key)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("images/astro.png")

plt.imshow(img)
```

```python
def image_to_text(image):

    model = genai.GenerativeModel(model_name="gemini-1.5-flash")

    prompt = "What is this image? Describe in one line."

    response = model.generate_content([img, prompt])

    image_text = response.text

    print(image_text)

    return image_text
```

```python
from gtts import gTTS

def text_to_speach(text):
    tts = gTTS(text, lang='en')
    file_path = "output/output.mp3"
    tts.save(file_path)
    return file_path
```

```python
def run_module(image):
    image_text = image_to_text(image)
    text_to_speach(image_text)
```

```python
run_module(img)
```
